```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
scaled_data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

```

```python
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(scaled_data)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(scaled_data)
df['Kmeans_Cluster'] = kmeans.labels_
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3)
df['Agglomerative_Cluster'] = agg.fit_predict(scaled_data)
```

```python
df['Cluster_Difference'] = df['Kmeans_Cluster'] - df['Agglomerative_Cluster']
```

```python
print(df['Cluster_Difference'].value_counts())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
```

```python
kmeans_pca = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans_pca.fit(principal_df)
```

```python
agg_pca = AgglomerativeClustering(n_clusters=3)
```

```python
df['PCA_Kmeans_Cluster'] = kmeans_pca.labels_
df['PCA_Agglomerative_Cluster'] = agg_pca.fit_predict(principal_df)
```

# **Inferences:**

Based on the analysis conducted on the dataset using K-means clustering, agglomerative clustering, and PCA, along with exploratory data analysis (EDA), the following inferences can be drawn:

1. **Exploratory Data Analysis (EDA) Inferences**:
   - The dataset contains information about customers' spending habits on various product categories.
   - There are no missing values in the dataset.
   - The distributions of spending on different product categories vary widely, indicating heterogeneity among customers.

2. **Data Preprocessing**:
   - Data preprocessing steps such as scaling may be required to ensure that features are on a similar scale, especially for clustering algorithms.

3. **Optimal Number of Clusters**:
   - The optimal number of clusters may be determined using techniques like the elbow method or silhouette score.

4. **K-means Clustering**:
   - K-means clustering partitions the customers into K clusters based on their spending patterns.
   - Each cluster represents a group of customers with similar spending behavior.
   - Cluster centroids represent the average spending pattern of customers within each cluster.

5. **Agglomerative Clustering**:
   - Agglomerative clustering creates a hierarchy of clusters by merging the most similar clusters iteratively.
   - The resulting dendrogram can be used to visualize the hierarchical structure of the clusters.

6. **Comparison of Cluster Memberships**:
   - The cluster memberships obtained from K-means and agglomerative clustering may show some differences due to the nature of the algorithms.

7. **Principal Component Analysis (PCA)**:
   - PCA reduces the dimensionality of the dataset while preserving most of the variance.
   - Principal components represent orthogonal directions in the feature space that capture the most variance in the data.

8. **Clustering with Principal Components**:
   - Clustering on principal components allows for visualization of high-dimensional data in a 2D space.
   - It may reveal underlying patterns or clusters that were not apparent in the original feature space.

9. **Overall Inferences**:
   - Customer segments or clusters identified through clustering algorithms can provide valuable insights for targeted marketing strategies.
   - PCA can help identify the most important features driving customer segmentation.
   - The choice of clustering algorithm and the number of clusters can significantly impact the interpretation of results.
